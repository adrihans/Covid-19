## Preprocessing of the sentences:

### Loading the datasets:

```python
df_sentences=pd.read_csv('cord19_df_sentences.csv')
```

```python
df_sentences.head()
```

```python
df_sentences.shape
```

```python
df_sentences_pre_10000=pd.read_csv('cord19_df_sentences_pre_10000.csv')
```

```python
df_sentences_pre_10000.head()
```

```python
df_sentences_indexed_pre_10000=pd.read_csv('cord19_df_sentences_pre_10000_indexed.csv')
```

```python
df_sentences_indexed_pre_10000.head()
```

```python
df_sentences_indexed_pre_10000.shape
```

```python
df_sentences_7000=pd.read_csv('cord19_df_sentences_7000.csv')
```

```python
df_sentences_7000.head()
```

```python
df_sentences_7000.shape
```

```python
df_post_7000=pd.read_csv('cord19_df_sentences_post_7000.csv')
```

```python
df_post_7000.head()
```

```python
df_post_7000.shape
```

```python
df_list_sentences_end=pd.read_csv('cord19_df_list_sentences_19327.csv')
```

```python
df_list_sentences_end.shape
```

```python
df_list_sentences_end.head()
```

### Constructing the dataframe:

```python
preprocessed_sentences=pd.concat((df_sentences_pre_10000['sentences_pre_10000'], df_post_7000['sentences_post_7000']),ignore_index=True)
```

```python
preprocessed_sentences.shape
```

```python
list_sentences=pd.concat((df_sentences_indexed_pre_10000['list_sentences'],  df_list_sentences_end['list_sentences']  ), ignore_index=True)
```

```python
list_sentences.shape
```

### Final DataFrame:

```python
d={'list_sentences': list_sentences, 'preprocessed_sentences': preprocessed_sentences}
df_preprocessed_indexed_sentences=pd.DataFrame(data=d)
```

```python
df_preprocessed_indexed_sentences.shape
```

```python
df_preprocessed_indexed_sentences.head()
```

```python
df_preprocessed_indexed_sentences.to_csv('df_covid_sentences_preprocessed_indexed.csv', index=False)
```
